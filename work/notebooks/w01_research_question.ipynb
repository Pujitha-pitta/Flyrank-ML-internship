## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*



1. Decision

Determine which search result pages are likely to have low click-through rates.

2. Action

SEO teams can prioritize pages predicted to have low CTR and improve them by updating titles, descriptions, keywords, or page content.

3. Cost of a Wrong Recommendation

If a page with high CTR is incorrectly predicted as low CTR, time and resources may be spent optimizing a page that already performs well.

If a page with low CTR is incorrectly predicted as high CTR, an opportunity to improve search performance and attract more users may be missed.
